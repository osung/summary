```python
dfs = tabula.read_pdf(pdf_file_path, pages='all', multiple_tables=True)
```

```python
stock_types = ['구매', '판매', '배당금입금', '외화증권배당금입금', '배당단주대금입금', '분배금', '소수점이벤트입고', '출석체크이벤트입고', '무상주입고', '유상청약대금출금']
```

```python
base_df = pd.DataFrame(columns=['티커', '종목명', '구분', '날짜', '금액', '수량', '총액', '증권사'])
profit_df = pd.DataFrame(columns=['티커', '종목명', '매수총액', '매수총수량', '평균매수가', '매도총액', '매도총수량', '평균매도가', '현재가', '평가손익', '배당수익', '실현손익', '총손익', '배당률', '수익률'])
```

```python
def split_code(name_code) :
    start_index = name_code.rfind('(')
    end_index = name_code.rfind(')')
    
    if start_index != -1 and end_index != -1:  # 괄호가 찾아졌는지 확인
        name = name_code[:start_index].strip()  # 괄호 앞의 문자열
        code = name_code[start_index + 1:end_index]  # 괄호 안의 문자열
    else :
        name = name_code
        code = ''
        
    return name, code
```

```python
def analize_table(df) :
    stock_df = pd.DataFrame(columns=['티커', '종목명', '구분', '날짜', '금액', '수량', '총액', '증권사'])

    for index, row in df.iterrows() :
        if row['거래구분'] in stock_types :
            #print(row['거래구분'], row['종목명(종목코드)'])
            name, code = split_code(row['종목명(종목코드)'])

            row_df = pd.DataFrame([{'티커': code, '종목명': name, '구분': row['거래구분'], '날짜': row['거래일자'], 
                      '금액': row['단가'], '수량': row['거래수량'], '총액': row['거래대금'], '증권사': '토스'}])
            
            stock_df = pd.concat([stock_df, row_df], ignore_index=True)
            
    return stock_df
```

```python
for df in dfs :
    stock_df = analize_table(df)
    #print(stock_df)
    base_df = pd.concat([base_df, stock_df], ignore_index=True)
```

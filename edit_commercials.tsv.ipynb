```python
def drop_row(df, index) :
    return df.drop(index=index)
```

```python
def add_to_final(final_df, f_df, df):
    final_df = final_df.append(f_df)
    df = df.drop(index = f_df.index)
    print('final:', len(final_df))
    print('df:', len(df))
    
    return final_df, df
```

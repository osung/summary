```python
for index, row in df.iterrows():
    print(f"인덱스: {index}")
    print(row['text'], row['PEST'], row['KSIC'], len(row['text']))
    user_input = input("해당 row를 파일로 저장하시겠습니까? (y/n): ")
    if user_input.lower() == 'y':
        # 파일 저장
        with open(file_name, 'a') as f:
            row = df.iloc[index:index+1]
            row.to_csv(file_name, sep='\t', index=False, mode='a', header=False)    
    else:
        print("파일 저장을 취소합니다.")
```
